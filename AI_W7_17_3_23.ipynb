```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn import linear_model
url='/content/50_Startups.csv'
data= pd.read_csv(url)
data.head()
data.hist()
plt.show()
correlation = data.corr()
print(correlation)
sns.heatmap(correlation,annot =True)
sns.pairplot(data)

#########################################################33
#do chinh sat
model = linear_model. LinearRegression()
model.fit(X=pd.DataFrame(data['R&D Spend']),y=data['Profit'])
print('He so Intercept la : ',model.intercept_)
print ('He so coeffient la : ', model.coef_)
model.score(X=pd.DataFrame(data['R&D Spend']),y=data['Profit'])
y_pred= model.predict(X=pd.DataFrame(data['R&D Spend']))

data.plot(kind='scatter',x='R&D Spend',
          y='Profit',figsize=(9,9),
          color='black',xlim=(0,200000))
plt.plot(data['R&D Spend'],y_pred,color='blue')


###########################################################
model = linear_model. LinearRegression()
X=pd.DataFrame([data['R&D Spend'],data['R&D Spend']**2,data['R&D Spend']**3]).T
model.fit ( X ,y=data['Profit'])
print ('Intercept cua mo hinh bac 3 :',model.intercept_)#mpg=-13.38 * wt + 1.17 wt^2 + 49.903
print('Coeff cua mo hinh bac 3 :',model.coef_)
print('Do chinh sat cua mo hinh bac 3 :',model.score(X,y=data['Profit']))

range = np.arange (0,200000 ,1)
X_range = pd.DataFrame ([range,range**2,range**3]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='R&D Spend',
          y='Profit',figsize=(9,9),
          color='black',xlim=(0,200000))
plt.plot(range,y_pred,color='blue')

###########################################################



model = linear_model. LinearRegression()
model.fit(X=data.loc[:,['R&D Spend','Administration','Marketing Spend']],y=data['Profit'])
print('Intercept la : ',model.intercept_)
print('coeffdent la : ',model.coef_)
print('do chinh sat la : ', model.score(X=data.loc[:,['R&D Spend','Administration','Marketing Spend']],y=data['Profit']))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn import linear_model
url='/content/Advertising.csv'
data= pd.read_csv(url)
data.head()
data.hist()
plt.show()
correlation = data.corr()
print(correlation)
sns.heatmap(correlation,annot =True)
sns.pairplot(data)
###################################################  BAC !
model = linear_model. LinearRegression()
model.fit(X=pd.DataFrame(data['TV']),y=data['sales'])
print('He so Intercept cua bac 1 la : ',model.intercept_)
print ('He so coeffient  cua bac 1 la : ', model.coef_)
model.score(X=pd.DataFrame(data['TV']),y=data['sales'])
y_pred= model.predict(X=pd.DataFrame(data['TV']))

data.plot(kind='scatter',x='TV',
          y='sales',figsize=(9,9),
          color='black',xlim=(0,400))
plt.plot(data['TV'],y_pred,color='blue')


###################################################
#Bac 5
#do chinh sat
model = linear_model. LinearRegression()
X=pd.DataFrame([data['TV'],data['TV']**2,data['TV']**3,data['TV']**4,data['TV']**5]).T
model.fit ( X ,y=data['sales'])
print ('Intercept cua mo hinh voi bac 5 :',model.intercept_)
print('Coeff cua mo dinh voi bac 5 :',model.coef_)
print('Do chinh sat cua mo hinh voi bac 5 :',model.score(X,y=data['sales']))

range = np.arange (0,300 ,0.1)
X_range = pd.DataFrame ([range,range**2,range**3,range**4,range**5]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='TV',
          y='sales',figsize=(9,9),
          color='black',xlim=(0,400))
plt.plot(range,y_pred,color='blue')


###################################################

model = linear_model. LinearRegression()
model.fit(X=data.loc[:,['TV','radio','newspaper']],y=data['sales'])
print('Intercept da bien la : ',model.intercept_)
print('Coeffdent da bien la : ',model.coef_)
print('Do chinh sat la : ', model.score(X=data.loc[:,['TV','radio','newspaper']],y=data['sales']))


```
